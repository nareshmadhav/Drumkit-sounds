Adding column to the data set

```python
df['total score']=df['math score']+df['reading score']+df['writing score']
df['average']=df['total score']/3
df.head()
```

The new column name is gender status idf the male candidate is there print eligible for outdoor games else elgible for indoor and outdoor games

```python
df['gender status']=['eligible for outdoor games' if x=='male' else 'eligible for indoor and outdoor games' for x in df['gender']]
df.head()
```

```python
# df['grade']=["A grade" if x>80 and x<=100 elif x>=50 and x<80 "B grade" for x in df['math score']]
# df['grade']=['A grade' if x>80 and x<=100 else 'B grade' if x>=50 and x<80 else 'C grade' for x in df['math score']]
# df.head()
```

```python
grade=[]
for x in df['math score']:
  if x>80 and x<=100:
    grade.append('A grade')
  elif x>=50 and x<80:
    grade.append('B grade')
  elif x>=35 and x<50:
    grade.append('C grade')
  else:
    grade.append('Fail')
df['math_grade']=grade
df.head()
```

```python
reading_full=df[df['reading score']==100]['average'].count()
writing_full=df[df['writing score']==100]['average'].count()
math_full=df[df['math score']==100]['average'].count()
print(f'Number of students with full marks in reading : {reading_full}')
print(f'Number of students with full marks in writing : {writing_full}')
print(f'Number of students with full marks in math : {math_full}')
```

```python
total_students=len(df)
percentage_math=(math_full/total_students)*100
percentage_reading=(reading_full/total_students)*100
percentage_writing=(writing_full/total_students)*100
print(f'Percentage of students with full marks in math : {percentage_math}')
print(f'Percentage of students with full marks in reading : {percentage_reading}')
print(f'Percentage of students with full marks in writing : {percentage_writing}')
```
